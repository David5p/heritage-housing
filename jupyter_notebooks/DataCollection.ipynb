## Install Python packages in the notebooks ##

### Change working directory ###

```python
from pathlib import Path
import os
import zipfile
import pandas as pd
```

```python
BASE_DIR = Path.cwd().resolve()
if "jupyter_notebooks" in str(BASE_DIR):
    BASE_DIR = BASE_DIR.parent

print("Project root:", BASE_DIR)
```

# Fetch data from Kaggle 

```python
KaggleDatasetPath = "codeinstitute/housing-prices-data"
DestinationFolder = BASE_DIR / "inputs/datasets/raw"
DestinationFolder.mkdir(parents=True, exist_ok=True)

# Download from Kaggle
os.environ["KAGGLE_CONFIG_DIR"] = str(BASE_DIR)

download_result = os.system(f'kaggle datasets download -d {KaggleDatasetPath} -p "{DestinationFolder}"')

zip_files = list(DestinationFolder.glob("*.zip"))

if zip_files:
    zip_path = zip_files[0]

    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(DestinationFolder)

    os.remove(zip_path)
    print("Kaggle dataset downloaded and extracted.")

else:
    print("Kaggle download failed (likely missing kaggle.json). Using existing dataset from outputs instead.")
```

# Load and Inspect Kaggle data

```python
# Load and Inspect Dataset

csv_files = list(DestinationFolder.rglob("house_prices_records.csv"))

if csv_files:
    df = pd.read_csv(csv_files[0])
    print("Dataset loaded successfully from extracted Kaggle files.")
else:
    raise FileNotFoundError("house_prices_records.csv not found in extraction folder.")

# Quick inspection
df.head()
df.info()

# Check duplicates
duplicates = df[df.duplicated()]
print(f"Number of duplicate rows: {len(duplicates)}")
```

# Save Clean Dataset to Collection Folder

```python
import os
os.makedirs('outputs/datasets/collection', exist_ok=True)

df.to_csv('outputs/datasets/collection/house_prices_records.csv', index=False)

```
